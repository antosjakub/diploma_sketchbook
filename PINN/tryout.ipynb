```python
from src import *
import torch
```

```python
class SIMPLE(nn.Module):
    def __init__(self, d, hidden_size=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d, hidden_size),
            nn.Linear(hidden_size, 1),
        )
    
    def forward(self, x):
        return self.net(x)
```

```python
d = 2
model = SIMPLE(d, 8)
```

```python
lr=1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
optimizer.zero_grad()

x = torch.rand(10, d, requires_grad=True)
u = model(x)
u
```

```python
grad_u = torch.autograd.grad(
    outputs=u,
    inputs=x,
    grad_outputs=torch.ones_like(u),
    create_graph=True,
)[0]
grad_u
```

```python
i = 0
torch.autograd.grad(
    outputs=grad_u[:,i].sum(),
    inputs=x,
    create_graph=True,
    allow_unused=True
)
```
