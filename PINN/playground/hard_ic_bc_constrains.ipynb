```python
import torch
```

```python
f = lambda x: x[:,0]*(1-x[:,0])*x[:,1]*(1-x[:,1])
```

```python
N = 100
device = 'cpu'
x = torch.linspace(0, 1, N, device=device)
y = torch.linspace(0, 1, N, device=device)
X_grid, Y_grid = torch.meshgrid(x, y, indexing='ij')
x_flat = X_grid.reshape(-1, 1)
y_flat = Y_grid.reshape(-1, 1)
X = torch.cat([x_flat, y_flat], dim=1)

f_vec = f(X)
F_grid = f_vec.reshape(N, N)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].contourf(X_grid, Y_grid, F_grid, levels=50, cmap='jet')
axes[0].set_title("")

axes[1].plot(x, x*(1-x))
axes[1].set_title("")
axes[1].set_ylim(0,1)
```
