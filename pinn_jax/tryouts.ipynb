```python
import jax.numpy as jnp
```

```python
dim = 3
k = jnp.ones(dim) * jnp.pi
params = {"alpha": 0.1, "k": k}
k_sq_sum = jnp.sum(k ** 2)

def exact_fn(t, x, p):
    return jnp.prod(jnp.sin(p["k"] * x)) * jnp.exp(-p["alpha"] * jnp.sum(p["k"] ** 2) * t)
```
