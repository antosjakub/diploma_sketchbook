```python
import numpy as np
```

```python
shape = (64**6,)
int(np.log10(shape[0]))
```

```python
np.ones(*shape, dtype=np.int8)
```

```python
# Create a memory-mapped file on disk
filename = "large_array.dat"
large_array = np.memmap(filename, dtype=np.int8, mode="w+", shape=shape)

# Fill it with data (only modified parts are loaded into RAM)
chuck_size = 64**5
for i in range(0, shape[0], chuck_size):
    large_array[i:(i+chuck_size)] = 1


# Access a small part (fast, doesn't load full array)
print(large_array[0:10])

# Flush changes to disk (important!)
large_array.flush()
```
