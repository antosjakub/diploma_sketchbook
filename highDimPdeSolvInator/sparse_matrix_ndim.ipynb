```python
def to_nd_from_1d(i):
    k = np.zeros([dim], dtype=int)
    i_rem = i
    for j in range(dim):
        kj = i_rem // N**(dim-(j+1))
        i_rem -= kj * N**(dim-(j+1))
        k[j] = kj
    return k
```

```python
indx_min = 0
indx_max = N-1
data_size = (2*dim+1)*n
data = np.zeros(data_size)
row_ind = np.zeros(data_size)
col_ind = np.zeros(data_size)
#B = np.zeros((n,n), dtype=int)
j=0
for i in range(n): # go ever each row of the stiffness matrix
    k = to_nd_from_1d(i)
    #B[i,i] = 2*dim
    data[j] = 2*dim
    row_ind[j] = i
    col_ind[j] = i
    j+=1
    for d in range(dim):
        if k[d]+1 <= indx_max:
            #B[i,i+N**(dim-(d+1))] = -1
            data[j] = 2*dim
            row_ind[j] = i
            col_ind[j] = i
        if k[d]-1 >= indx_min:
            #B[i,i-N**(dim-(d+1))] = -1
            data[j] = 2*dim
            row_ind[j] = i
            col_ind[j] = i
```
